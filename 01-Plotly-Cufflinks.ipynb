#### check the differnce of using plotly, by using two different methods ".plot()" and ".iplot()" 

```python
df1.plot(figsize=(14, 4))
```

```python
df1.iplot(size=(30, 4))

# you can notice the changes between both plots 
```

### different types of plots 

#### scatter plot 

```python
# using scatter plot between culomns a, b

df1.iplot(kind='scatter', x='a', y='b')
```

```python
# scatter plot with arguments 

df1.iplot(kind='scatter', x='a', y='b', mode='markers',size=10, color = 'lightblue')
```

#### bar plot 

```python
# using df2

df2.iplot(kind='bar', x='Category', y='Values', 
          title='Bar Plot',
          xTitle='Category',
          yTitle='Values',
          barmode='group',   # Set the bar mode (group, stack, overlay)
          bargap=0.2,  # Set the gap between bars
          bargroupgap=0.1,  # Set the gap between bar groups
          showlegend=True,  # Show or hide the legend
         )
     
```

```python
# using big data like df1, and use grouping methods  

df1.count().iplot(kind='bar')
```

```python
df1.sum().iplot(kind='bar')
```

```python
df1.mean().iplot(kind='bar')
```

#### box plot 

```python
df1.iplot(kind='box')
```

#### 3D surface plot 

```python
# calling 3D surface plot

df1.iplot(kind='surface')
```

```python
# using some arguments 

df1.iplot(kind='surface', colorscale='rdylbu', color='coolwarm', width=0.5)
```

#### histogram plot 

```python
# calling one column

df1['a'].iplot(kind='hist')
```

```python
# adding some arguments 

df1['a'].iplot(kind='hist', bins=50)
```

```python
# calling the whole dataset

df1.iplot(kind='hist', bins=60)
```

#### spread 

```python
df1[['a','b']].iplot(kind='spread')
```

#### bubble plot

```python
# comparing two columns 'a,b' to column 'c'

df1.iplot(kind='bubble', x='a', y='b', size='c',  color= 'slateblue')
```

#### scatter matrix

```python
df1.scatter_matrix()
```
